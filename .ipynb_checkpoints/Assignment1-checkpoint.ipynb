Task 1: Feature Generation (15%)
Use and run the following code (a deep network) to generate features from a set of training images. For this assignment, you do not need to know how the deep network is working here to extract features.
This code extracts the features of image T4.tif (in the T folder of dataset). Modify the code so that it iterates over all images of the dataset and extracts their features.

Allocate 10% of the data for validation.

Insert your code here for Task 1





```python
import os
import random
import numpy as np
import torch
import torchvision.transforms as transforms
from torchvision.models import densenet121
from torch.autograd import Variable
from PIL import Image

from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings("ignore")

# Load pre-trained DenseNet model
model = densenet121(pretrained=True)

# Remove the classification layer (last fully connected layer)
model = torch.nn.Sequential(*list(model.children())[:-1])

# Add a global average pooling layer
model.add_module('global_avg_pool', torch.nn.AdaptiveAvgPool2d(1))

# Set the model to evaluation mode
model.eval()

# Define the image preprocessing pipeline
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

def extract_features(image_path):
    # Load an image
    image = Image.open(image_path)

    # Preprocess the image
    input_tensor = preprocess(image)
    input_batch = input_tensor.unsqueeze(0)  # Add a batch dimension

    # Wrap the input tensor in a Variable
    input_var = Variable(input_batch)

    # Forward pass through the model
    features = model(input_var)

    # Extract the feature tensor
    feature_vector = features.squeeze().detach().numpy()

    return feature_vector

# Walk the directory structure and find all folders
imageDirectory = sorted(list(os.listdir('data')))

# shuffle the data
random.shuffle(imageDirectory)

labels = []
features = []

for dir in imageDirectory:
    img_path = 'data/' + dir

    # Every folder has 39 images
    imgs = sorted(list(os.listdir(img_path)))

    for i, img in enumerate(imgs):

        # Split into training and test
        feature = extract_features( os.path.join(img_path, img) )
        features.append(feature)
        labels.append(dir)

# Split the data into training and test sets (test size of 10%)
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.1, random_state=42)
```

Task 2: High Bias Classification Method (5%)
Choose a classification method and let is have a `high bias`.
Train it on the generated features and discuss why it is underfitting.

Insert your code here for Task 2




```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

# Initialize and train a Linear SVM with high bias
svm_model = LinearSVC(C=0.0000001)
svm_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("[RESULT] Accuracy: {:.2f}%".format(accuracy*100))
```

We trained a linear SVM with a high bias for this task. C, the regularization parameter was set to an extremely low value of 0.0000001  By choosing a small regularization value, we are decreasing the importance of misclassifying points in the model (decreasing the penalty of misclassification). This makes the model less sensitive to the training data, meaning that it underfits.

Task 3: High Variance Classification Method (5%)
Use the chosen classification method and let it have a `high variance`.
Train it on the generated features and discuss why it is overfitting.

Insert your code here for Task 3




```python
from sklearn import svm

# Initialize and train a SVM with high variance
svm_model = svm.SVC(gamma=0.05)
svm_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("[RESULT] Accuracy: {:.2f}%".format(accuracy*100))
```

We trained an SVM (with rbf kernel) with high variance for this task. The regularization parameter was left as the default value (1.0), while the kernel coefficient () was set to a comparatively small value of 0.05. The kernel coefficient is inversely proportional to the variance of the RBF (Gaussian) kernel. This means that points which are farther apart can still be considered “similar” by the kernel, leading to increased sensitivity to training data and noise, leading to overfitting.

Task 4: Balanced Classification Method (15%)
Use the chosen classification method and let it `balance the bias and variance`.
Train it on the generated features, possibly adjusting parameters.
Discuss insights into achieving balance.

Insert your code here for Task 4




```python
# Apply GridSearchCV to find the best hyperparameters
from sklearn.model_selection import GridSearchCV
import numpy as np

NUM_CPUS=-1

def gridSearch(estimator, param_grid, X_train, y_train):
    grid = GridSearchCV(estimator=estimator, param_grid=param_grid, n_jobs=NUM_CPUS)
    grid.fit(X_train, y_train)

    # show the grid search information
    print("[INFO] Grid search best score: {:.2f}%".format(grid.best_score_ * 100))
    print("[INFO] Grid search best parameters: {}".format(grid.best_params_))

    return grid.best_params_, grid.best_score_

# Define the hyperparameters to search
# NOTE: RBF seems to perform the best, other kernels may have higher grid scores but seem to produce a worse accuracy
parameters = [
	# {"kernel":
	# 	["linear"],
	# 	"C": [0.0001, 0.001, 0.1, 1, 10, 100, 1000]},
	# {"kernel":
	# 	["poly"],
	# 	"degree": [2, 3, 4],
	# 	"C": [0.0001, 0.001, 0.1, 1, 10, 100, 1000]},
	{"kernel":
		["rbf"],
		"gamma": [0.0001, 0.001, 0.1, 1, 10, 100, 1000],
		"C": [0.0001, 0.001, 0.1, 1, 10, 100, 1000]}
]

print("[INFO] Searching for the best hyperparameters")
grid_best_params, grid_score = gridSearch(svm.SVC(), parameters, X_train, y_train)

# Refine the search space once
print("[INFO] Refining the search space")
C_range = np.linspace(grid_best_params['C'] * 0.5, grid_best_params['C'] * 1.5, num=10)

if grid_best_params['kernel'] == 'rbf':
	gamma_range = np.linspace(grid_best_params['gamma'] * 0.5, grid_best_params['gamma'] * 1.5, num=10)
	kernel_range = ["rbf"]
	grid_params = dict(kernel=kernel_range, C=C_range, gamma=gamma_range)

	grid_best_params, grid_score = gridSearch(svm.SVC(), grid_params, X_train, y_train)

elif grid_best_params['kernel'] == 'linear':
	kernel_range = ["linear"]
	grid_params = dict(kernel=kernel_range, C=C_range)

	grid_best_params, grid_score = gridSearch(svm.SVC(), grid_params, X_train, y_train)

elif grid_best_params['kernel'] == 'poly':
	degree_range = [2, 3, 4]
	kernel_range = ["poly"]
	grid_params = dict(kernel=kernel_range, C=C_range, degree=degree_range)

	grid_best_params, grid_score = gridSearch(svm.SVC(), grid_params, X_train, y_train)


# Initialize and train a SVM with the best hyperparameters
clf = svm.SVC(**grid_best_params)
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("[RESULT] Accuracy: {:.4f}% achieved with C: {:.4f}, degree: {}, gamma: {:.6f}, and kernel: {}".format(accuracy*100, clf.C, clf.degree, clf.gamma, clf.kernel))
```

By leveraging GridSearch, we were able to tune various hyperparameters across specified ranges to balance bias and variance. Based on what kernel we chose, we would provide a different range of values for the kernel-specific parameters. From the findings, it was found that an RBF kernel produced good accuracy (96.15%) with a C value of 8.3 and a gamma of 0.00117.

In this case, a balance between bias and variance was achieved via proper tuning of the SVM hyperparameters (C and gamma). For this particular dataset, a high bias and low variance model performs well. This makes sense since the model with high bias still performed relatively well (64.1%), while the model with high variance performed very poorly (20.51%).

Task 5: K-Means Clustering (20%)
Apply `K-Means` clustering on the generated features.
Test with available labels and report accuracy.
Experiment with `automated K` and compare with `manually set 20 clusters`.

Insert your code here for Task 5




```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score, adjusted_mutual_info_score, homogeneity_score, completeness_score, v_measure_score, v_measure_score, normalized_mutual_info_score

def calc_silhouette(data, k):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    labels = kmeans.predict(data)
    return silhouette_score(data, labels)

# Calculate the silhouette score for a range of cluster sizes (2 to 25)
k_values = range(3, 25)
silhouette_scores = [calc_silhouette(X_train, k) for k in k_values]

# Find the best k
best_k = k_values[np.argmax(silhouette_scores)]

print("[INFO] Best k: {}".format(best_k))

# Fit the KMeans model on the training data with the best k
kmeans = KMeans(n_clusters=best_k, random_state=42)
kmeans.fit(X_train)

# Predict the labels of the test set
y_pred = kmeans.predict(X_test)

print("[INFO] Calculating the accuracy of the KMeans model with k={}".format(best_k))

# Calculate accuracy
nmi = normalized_mutual_info_score(y_test, y_pred)
ari = adjusted_rand_score(y_test, y_pred)
ami = adjusted_mutual_info_score(y_test, y_pred)
homogeneity = homogeneity_score(y_test, y_pred)
completeness = completeness_score(y_test, y_pred)
v_measure = v_measure_score(y_test, y_pred)

print("[RESULT] Normalized Mutual Information (NMI): {:.4f}%".format(nmi*100))
print("[RESULT] Adjusted Rand Index (ARI): {:.4f}%".format(ari*100))
print("[RESULT] Adjusted Mutual Information (AMI): {:.4f}%".format(ami*100))
print("[RESULT] Homogeneity: {:.4f}%".format(homogeneity*100))
print("[RESULT] Completeness: {:.4f}%".format(completeness*100))
print("[RESULT] V-measure: {:.4f}%".format(v_measure*100))


# Fit the KMeans model on the training data with k=20 (20 classes total)
kmeans = KMeans(n_clusters=20, random_state=42)
kmeans.fit(X_train)

# Predict the labels of the test set
y_pred = kmeans.predict(X_test)

print("\n[INFO] Calculating the accuracy of the KMeans model with k={}".format(20))

# Calculate accuracy
nmi = normalized_mutual_info_score(y_test, y_pred)
ari = adjusted_rand_score(y_test, y_pred)
ami = adjusted_mutual_info_score(y_test, y_pred)

homogeneity = homogeneity_score(y_test, y_pred)
completeness = completeness_score(y_test, y_pred)
v_measure = v_measure_score(y_test, y_pred)

print("[RESULT] Normalized Mutual Information (NMI): {:.4f}%".format(nmi*100))
print("[RESULT] Adjusted Rand Index (ARI): {:.4f}%".format(ari*100))
print("[RESULT] Adjusted Mutual Information (AMI): {:.4f}%".format(ami*100))
print("[RESULT] Homogeneity: {:.4f}%".format(homogeneity*100))
print("[RESULT] Completeness: {:.4f}%".format(completeness*100))
print("[RESULT] V-measure: {:.4f}%".format(v_measure*100))
```

For the automated K, values between 3 and 25 were tested. The K with the highest silhouette_score was selected. A value of K = 8 was obtained.

As expected K=8 performs worse than K=20 in every measured metric that was used. Some metrics include adjusted rand index (ARI), normalized mutual information (NMI), and adjusted mutual information (AMI). The adjusted rand index (ARI) compares the number of correctly predicted pairs with the total number of possible pairs while rejecting chance by discounting a chance normalization term. The ARIs for k=8 and k= 20 are 19.9184% and 61.7561% respectively. The mutual information measures the agreement between the cluster assignments using joint and marginal probabilities, and has 2 variants in normalized and adjusted mutual information (NMI and AMI). The NMI is the mutual information divided by average cluster entropies, and the AMI is the NMI adjusted for chance by discounting a chance normalization term. The NMI for k=8 and k= 20 are 66.6876% and 86.7094% respectively, and the AMI for k=8 and k= 20 are 51.2942%% and 72.8576% respectively. 

From the dataset, we already know that there are 20 classes (labels). As such, it is expected that K=20 will perform the best.

Task 7: PCA for Classification Improvement (20%)
Apply `PCA` on the features and then feed them to the best classification method in the above tasks.
Assess if PCA improves outcomes and discuss the results.

Insert your code here for Task 7




```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score, homogeneity_score, completeness_score, v_measure_score, v_measure_score, normalized_mutual_info_score

######## PCA ########
variance = 0.95
pca = PCA(variance, random_state=42)
pca.fit(X_train)
########################

# Reduce the dimensionality of the training and test data
X_train_reduced = pca.transform(X_train)
X_test_reduced = pca.transform(X_test)

# Compare number of components before and after PCA
print("[INFO] Original number of features: {}".format(np.array(X_train).shape[1]))
print("[INFO] Reduced number of features: {}".format(X_train_reduced.shape[1]))

# Fit the KMeans model on the training data with k=20 (20 classes total)
kmeans = KMeans(n_clusters=20, random_state=42)
kmeans.fit(X_train_reduced)

# Predict the labels of the test set
y_pred = kmeans.predict(X_test_reduced)

print("\n[INFO] Calculating the accuracy of the KMeans model with k={} after using PCA".format(20))

# Calculate accuracy
nmi = normalized_mutual_info_score(y_test, y_pred)
ari = adjusted_rand_score(y_test, y_pred)
ami = adjusted_mutual_info_score(y_test, y_pred)

homogeneity = homogeneity_score(y_test, y_pred)
completeness = completeness_score(y_test, y_pred)
v_measure = v_measure_score(y_test, y_pred)

print("[RESULT] Normalized Mutual Information (NMI): {:.4f}%".format(nmi*100))
print("[RESULT] Adjusted Rand Index (ARI): {:.4f}%".format(ari*100))
print("[RESULT] Adjusted Mutual Information (AMI): {:.4f}%".format(ami*100))
print("[RESULT] Homogeneity: {:.4f}%".format(homogeneity*100))
print("[RESULT] Completeness: {:.4f}%".format(completeness*100))
print("[RESULT] V-measure: {:.4f}%".format(v_measure*100))
```

When comparing the resulting accuracy of the KMeans model with k=20 against the same metrics from Task 5, we see that there is a slight dropoff after performing PCA. This is because PCA treats features with large variance as important but that can result in features that are not as useful being retained while other more useful features for our task are dropped. 

PCA reduced the dimensionality of the data from 1024 features down to 127. While this will undoubtedly improve computation speed, such a significant reduction of features may have caused us to lose some information that is important for classification.

Task 8: Visualization and Analysis (10%)
`Plot the features` in a lower dimension using dimentinality reduction techniques.
Analyze the visual representation, identifying patterns or insights.

Insert your code here for Task 8

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

######## PCA ########
pca = PCA(n_components=2, random_state=42)
pca.fit(X_train)
########################

# Reduce the dimensionality of the training and test data
X_train_reduced = pca.transform(X_train)
X_test_reduced = pca.transform(X_test)

# Convert categorical labels to numeric values using LabelEncoder
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)



# Visualize PCA components of Training data
plt.figure(figsize=(10, 6))

# Plot scatter plot with numeric labels
plt.scatter(X_train_reduced[:, 0], X_train_reduced[:, 1], c=y_train_encoded, cmap='tab20', s=30, alpha=0.6)

plt.title('PCA Components of Training Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Classes', ticks=range(20))
plt.show()

# Visualize PCA components of Test data
plt.figure(figsize=(10, 6))

# Plot scatter plot with numeric labels
plt.scatter(X_test_reduced[:, 0], X_test_reduced[:, 1], c=y_test_encoded, cmap='tab20', s=30, alpha=0.6)

plt.title('PCA Components of Test Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Classes', ticks=range(20))
plt.show()
```

We chose to plot the training and test data into score plots based on the two principal components after PCA. This allows us to gain a quick visual representation of how each component influences our model and where classes overlap or differ. A further calculation would allow us to know how much information each component holds and how much information was lost projecting down to two dimensions.

With 20 different classes, visualizing all at once can become quite difficult. However, some observations that we can make are:
Although the classes are not linearly separable via the 2 principal components shown, clusters are still able to be seen in 2D (although many of them overlap). This implies that they are likely able to be (at least partially) linearly separated in higher dimensions. The dimensionality was reduced from 1024 to 2, which is a significant difference. Additionally, the high performance of K-means clustering in task 5 also supports this theory, since K-means is good at clustering linearly separable points.
Class two and nine are linearly separable from the rest to an extent based on principal component 1, with class two being more spread out on principal component 2
